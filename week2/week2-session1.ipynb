## 【問題1】1次関数
$ x $が-50から50の範囲で次の式を考えます。  
  
$$
    y = \frac{1}{2}x+1
$$
  
-50から50まで0.1間隔のxを表すndarrayと、各要素に対するyを表すndarrayを作成してください。  

```python
# coding utf8
# 【問題1】1次関数
import numpy  as np
import matplotlib.pyplot as plt

a = 0.5 # 定数0.5を変数に格納
x = np.arange(-50, 50.1, 0.1)# -50から50までの配列
y = (a * x) + 1

#print(type(y))
# print(x.shape)
print(y.shape)
print(y)
```

## 【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。0列目にx、1列目にyが来るようにします。

```python
ndarray_x_y = np.hstack((x.reshape(1001,1), y.reshape(1001,1)))
print(ndarray_x_y)
print(ndarray_x_y.shape)
print(ndarray_x_y.ndim)
```

## 【問題3】勾配を求める
各x座標における勾配を求めます。ここで勾配はyの変化量をxの変化量で割ることで計算します。  
各変化量は隣同士の要素の差から近似的に求めます。  
注意  
・厳密に各点における勾配を求めるには極限（微分）を利用する必要があります  
・勾配はx座標の間隔を使うので、xの要素数に対して一つ少なくなります  

```python
# 【問題3】勾配を求める
# 変化量d
d = ndarray_x_y[1:] - ndarray_x_y[:-1]
# 勾配s
s = d[:, 1] / d[:,0]
print(s)
```

```python
# 【問題5】Pythonの関数化

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
    np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
    xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
    関数の勾配。差をとるため要素は1つ減る
    """
    # print(*x_range)
    array_x = np.arange(*x_range)
    # shepeを(1000,1)に変更しておく
    #print(array_x.shape)
    array_x.shape = (*array_x.shape, 1)
    #print(array_x.shape)
    # 結合を行う、勾配を求める
    array_xy = np.hstack((array_x, function(array_x)))
    # 変化量delta
    delta = array_xy[1:] - array_xy[:-1]
    # 勾配gradient
    gradient = delta[:, 1] / delta[:,0]
    #print(gradient.shape)
    return array_xy, gradient


def function1(array_x):
    """
     y = x**2の関数。xの配列を引数とする。
     
    Parameters
    ----------------
    array_x : ndarray
    xの値を取りうる配列。

    Returns
    ----------------
    array_y : ndarray, shape(n, )
    y=x**2のyの値。nはarray_xに依存する。
    """
    array_y = array_x**2
    return array_y

def function2(array_x):
    """
    y = 2*x**2 + 2^xの関数。xの配列を引数とする。

    Parameters
    ----------------
    array_x : ndarray
    xの値を取りうる配列。

    Returns
    ----------------
    array_y : ndarray, shape(n, )
    y=2*x**2+2^xのyの値。nはarray_xに依存する。
    """
    array_y = 2*array_x**2 + 2**array_x
    return array_y

def function3(array_x):
    """
    y = sin(x**(1/2))の関数。xの配列を引数とする。

    Parameters
    ----------------
    array_x : ndarray
    xの値を取りうる配列。

    Returns
    ----------------
    array_y : ndarray, shape(n, )
    y=2*x**2+2^xのyの値。nはarray_xに依存する。
    """
    import numpy as np
    array_y = np.sin(array_x**(1/2))
    return array_y
```

```python
array_xy, gradient = compute_gradient(function1)

plt.figure(dpi=90)
plt.title("y = x**2")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(array_xy.T[0], array_xy.T[1], label="y = x**2")
plt.legend()
plt.show()

plt.figure(dpi=90)
plt.title("y'")
plt.xlabel("x")
plt.ylabel("y'")
plt.plot(array_xy.T[0][1:],gradient, label="gradient")
plt.legend()
plt.show()
```

# 【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。  
今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。  
また、最小のyをとる前後の勾配を表示してください。  
[numpy.ndarray.min — NumPy v1.16 Manual](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.min.html)  
[numpy.ndarray.argmin — NumPy v1.16 Manual](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.argmin.html)  

```python
# 【問題6】最小値を求める

def min_index(function, x_range=(-50, 50.1, 0.1)):
    array_xy, gradient = compute_gradient(function, x_range)
    #print(array_xy.T[1][500])
    y_min =(array_xy.T[1].argmin(), array_xy.T[1].min())
    #print(y_min)
    if y_min[0] != 0:
        delta_y_min = (gradient[y_min[0] -1] , gradient[y_min[0]] )
    elif y_min[0] == 0:
        delta_y_min = (gradient[y_min[0]],)
    return y_min, delta_y_min
```
