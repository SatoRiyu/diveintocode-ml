## ダミーデータの作成
機械学習に関するプログラムの実験を行う上で、実データではないランダムなデータ（ダミーデータ）が欲しいということがある。正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていく。  

### 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成する。 
  
$$
        \left[\begin{array}{c}
            1.0 & 0.8 \\
            0.8 & 1.0\\
        \end{array}\right] 
$$
  
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成出来る。  
[numpy.random.multivariate_normal — NumPy v1.16 Manual](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.multivariate_normal.html)    
  
作成したndarrayのshapeは(500, 2)になる。 
  
シード値について  
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておく。今回シード値は0を使うことにする。
>np.random.seed(0)

```python
# coding utf8
# 
# 【問題1】乱数の作成

# モジュールのインポート
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 乱数のシード値を固定
np.random.seed(0)
# 定数の定義
MU = (-3, 0)  # 平均値mu
SIGMA = np.array([[1.0, 0.8], [0.8, 1.0]])  # 共分散行列sigma

# 正規分布による乱数作成
arry_random_number = np.random.multivariate_normal(MU, SIGMA, 500)
print("作成された乱数:{}".format(arry_random_number))
print("shape:{}".format(arry_random_number.shape))
```

#### 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化する。  
散布図はmatplotlibのplt.scatter()を使うことで描ける。  
[matplotlib.pyplot.scatter — Matplotlib 3.1.1 documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
# 【問題2】散布図による可視化


x = arry_random_number[:, 0]
y = arry_random_number[:, 1]

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x, y)

plt.show()
```

### 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化する。  
ヒストグラムはplt.hist()を使うことで描ける。  
[matplotlib.pyplot.hist — Matplotlib 3.1.1 documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html)  
各次元ごとに作成し、2つのヒストグラムとする。  
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにする。  
[matplotlib.pyplot.xlim — Matplotlib 3.1.1 documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.xlim.html)  

```python
# 乱数を一次元の配列にする
arry_random_number_1000 = arry_random_number.reshape(arry_random_number.shape)
# xlimに最小値、最大値を設定する
plt.xlim(np.min(arry_random_number_1000), np.max(arry_random_number_1000))
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(x,40)
plt.show()

plt.xlim(np.min(arry_random_number_1000), np.max(arry_random_number_1000))
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(y,40)
plt.show()
```

### 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成する。  
  
$$
        \left[\begin{array}{c}
            1.0 & 0.8 \\
            0.8 & 1.0\\
        \end{array}\right] 
$$
  
そして、問題1、4それぞれのデータをひとつの散布図として可視化する。凡例として問題1のものは0、問題2のものは1を表示する。

```python
# 【問題4】データの追加

# 定数の定義
MU_1 = (0, -3)  # 平均値mu
SIGMA = np.array([[1.0, 0.8], [0.8, 1.0]])  # 共分散行列sigma

# 正規分布による乱数作成
arry_random_number_1 = np.random.multivariate_normal(MU_1, SIGMA, 500)
print("作成された乱数:{}".format(arry_random_number_1))
print("shape:{}".format(arry_random_number_1.shape))
```

```python
# 問題1、4それぞれのデータをひとつの散布図として可視化
x_1 = arry_random_number_1[:, 0]
y_1 = arry_random_number_1[:, 1]

plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(x, y, label=0)
plt.scatter(x_1, y_1, label=1)
plt.legend()
plt.show()
```

### 【問題5】データの結合
データはまとめておいた方が後々扱いやすい。問題1、4で作成したndarrayを結合し、(1000, 2)のndarrayとする。結合はnp.concatenate()やnp.vstack()を使うことで行える。  
[numpy.concatenate — NumPy v1.16 Manual](https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html)  
[numpy.vsplit — NumPy v1.16 Manual](https://docs.scipy.org/doc/numpy/reference/generated/numpy.vsplit.html#numpy.vsplit)  

```python
arry_random_number_0_and_1 = np.vstack((arry_random_number, arry_random_number_1))
print("結合したndarray:{}".format(arry_random_number_0_and_1))
print("shape:{}".format(arry_random_number_0_and_1.shape))
```

### 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示すラベル付けを行う。問題1のものには0、問題4のものには1を対応させる。   
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成する。   
機械学習に使用するデータセットはこのような形になっていることが多い。   

```python
# メモリ領域の確保
new_arry_1000 = np.empty((1000, 3))
#print(new_arry_1000)

# 0のみの行列を作成
# shapeは(500,1)
arry_zeros = np.zeros((500,1))
#print(arry_zeros)

# 1のみの行列を作成
# shapeは(500,1)
arry_ones = np.ones((500,1))
#print(arry_ones)

# 0の行列と１の行列を縦に結合する
arry_zeros_ones = np.vstack((arry_zeros, arry_ones))
#print(arry_zeros_ones)
#print(arry_zeros_ones.shape)

# 乱数の行列に０と１の行列を横に結合する
new_arry_1000 = np.hstack((arry_random_number_0_and_1, arry_zeros_ones))

print("ラベル付けしたndarray:{}".format(new_arry_1000))
print("shape:{}".format(new_arry_1000.shape))
```
